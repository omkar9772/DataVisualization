Box Plot


```python
for column in df.columns[:-1]:
    sns.boxplot(x = 'Survived', y = column, data = df)
    plt.title("haberman data")
    plt.show()
```

There are 0 deaths before age 34 and there are 0 survivors after age 78
. there are many outliers in Positive lymph Nodes
. 50th percentile of Survived and not survied is same i.e. 1963

**violinplot**

```python
for column in df.columns[:-1]:
    sns.violinplot(x = 'Survived', y = column, data = df)
    plt.title("haberman data")
    plt.show()
```
